## Step 3: Load Data

```python
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

## Step 4: Clean Data

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

df = df.dropna()
df = df.drop_duplicates()

df = df.sort_values('Order Date')
```

## Step 5: Create Time-Series Data

```python
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()
daily_sales.set_index('Order Date', inplace=True)
```

## Step 6: Feature Engineering

```python
daily_sales['day'] = daily_sales.index.day
daily_sales['month'] = daily_sales.index.month
daily_sales['year'] = daily_sales.index.year
daily_sales['weekday'] = daily_sales.index.weekday

daily_sales['lag_1'] = daily_sales['Sales'].shift(1)
daily_sales = daily_sales.dropna()
```

## Step 7: Train-Test Split

```python
train = daily_sales[:'2017']
test = daily_sales['2018':]
```

## Step 8: Train Model

```python
from sklearn.ensemble import RandomForestRegressor

X_train = train.drop('Sales', axis=1)
y_train = train['Sales']

X_test = test.drop('Sales', axis=1)
y_test = test['Sales']

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

## Step 9: Predict

```python
test['Predicted'] = model.predict(X_test)
```

## Step 10: Export for Tableau

```python
test.reset_index(inplace=True)
test.to_csv('final_forecast.csv', index=False)
```
